# 00 Explore the raw data

Quick sanity check on the JSON in the UC volume: does it **read**, does the
**schema** match expectations (especially for the nested deck arrays), is the **count**
sensible, and does the **data** look real. Read-only.

```python
RAW = "/Volumes/workspace/clash/raw"
```

## Battles

`battles.jsonl` is newline-delimited JSON, one per ingestion-batch folder.

```python
battles = spark.read.json(f"{RAW}/*/battles.jsonl")
battles.printSchema()
```

```python
print(battles.count(), "battle rows")
```

```python
display(battles.limit(10))
```

## Players

A single pretty-printed JSON array, needs `multiline`.

```python
players = spark.read.option("multiline", "true").json(f"{RAW}/players.json")
players.printSchema()
print(players.count(), "players")
display(players.limit(10))
```

## Cards

```python
cards = spark.read.option("multiline", "true").json(f"{RAW}/cards.json")
cards.printSchema()
print(cards.count(), "cards")
display(cards.orderBy("card_id").limit(10))
```
